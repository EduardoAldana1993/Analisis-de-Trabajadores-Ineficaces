## Glosario:

En este punto estara el glosario actualizado de los terminos para el trabajo, se actualizara de ser necesario. Adicional se expondra uno en estos momentos.

**1. Análisis exploratorio de datos (EDA)**

Proceso inicial del análisis donde se examina el comportamiento de las variables mediante visualizaciones y estadísticas descriptivas. Sirve para identificar patrones, outliers y relaciones entre variables.

**2. Atípicos / Outliers**

Datos que se encuentran muy alejados del comportamiento normal del conjunto. Pueden indicar errores de captura o situaciones excepcionales que requieren análisis.

**3. Criterios de eficiencia**

Reglas cuantitativas definidas para clasificar a un operador como eficiente o ineficiente. Pueden basarse en percentiles, umbrales o métricas del negocio.

**4. Dataset**

Conjunto de datos estructurado, generalmente en formato tabla, que contiene la información a analizar en el proyecto.

**5. Hipótesis nula (H0)**

Planteamiento estadístico que afirma que no existe diferencia entre dos grupos o condiciones (ejemplo: operadores eficientes e ineficientes).

**6. Hipótesis alternativa (H1)**

Planteamiento contrario a la hipótesis nula. Afirma que sí existe una diferencia significativa entre los grupos analizados.

**7. Imputación**

Método para reemplazar valores faltantes en un dataset usando medias, medianas u otros criterios.

**8. Interpretación**

Explicación de lo que muestran las gráficas o resultados del análisis, traduciendo lo visual/numérico en conclusiones entendibles.

**9. Mann-Whitney U**

Prueba estadística no paramétrica usada para comparar dos grupos cuando los datos no siguen una distribución normal.

**10. Métrica**

Valor numérico que mide algún aspecto del desempeño de un operador, como tasa de pérdida, velocidad, productividad o resolución.

**11. Normalización**

Proceso para ajustar valores numéricos a un mismo rango o escala con el fin de facilitar comparaciones.

**12. Operador ineficiente**

Operador que, según los criterios definidos, presenta un desempeño inferior al esperado o por debajo de los umbrales establecidos.

**13. Preprocesamiento**

Etapa de limpieza y preparación del dataset antes de hacer análisis. Incluye manejo de nulos, corrección de errores y creación de nuevas métricas.

**14. Prueba estadística**

Método utilizado para determinar si una diferencia entre grupos es significativa o se debe al azar.

**15. Prueba t (t-test)**

Prueba estadística paramétrica que compara medias entre dos grupos cuando los datos siguen distribución normal.

**16. Segmentación**

Proceso de dividir a los operadores en grupos (por ejemplo, eficientes vs. ineficientes) según criterios definidos.

**17. Significancia estadística**

Indica si un resultado es suficientemente fuerte como para rechazar la hipótesis nula. Se evalúa con el valor p.

**18. Valor p**

Probabilidad de obtener un resultado igual o más extremo que el observado si la hipótesis nula fuera verdadera. Si p es menor a un umbral (por ejemplo, 0.05), se considera significativo.
